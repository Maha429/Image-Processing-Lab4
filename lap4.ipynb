```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

from skimage import data, img_as_float
from skimage import exposure

# Load moon image
img = data.moon()
p2, p98 = np.percentile(img, (2, 98))
img_rescale = exposure.rescale_intensity(img, in_range=(p2, p98))
plt.figure(figsize=(10,7))

plt.subplot(2,1,1)
plt.imshow(img, cmap='gray')
plt.title('Original image')
plt.axis('off')

plt.subplot(2,1,2)
plt.hist(img.flat, bins=256, range=(0,255))
plt.xlabel('gray level values')
plt.ylabel('Number of pixels')

plt.show()
```

```
p3, p80 = np.percentile(img, (3, 80))
img_rescale2 = exposure.rescale_intensity(img, in_range=(p3, p80))

plt.figure(figsize=(10,7))

plt.subplot(2,1,1)
plt.imshow(img_rescale2, cmap='gray')
plt.title('Contrast stretch (3%, 80%)')
plt.axis('off')

plt.subplot(2,1,2)
plt.hist(img_rescale2.flat, bins=256, range=(0,255))
plt.xlabel('gray level values')
plt.ylabel('Number of pixels')

plt.show()
```

```
# Histogram Equalization
img_eq = exposure.equalize_hist(img)

plt.figure(figsize=(10,7))

plt.subplot(2,1,1)
plt.imshow(img_eq, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.subplot(2,1,2)
plt.hist(img_eq.flat, bins=256)
plt.xlabel('gray level values')
plt.ylabel('Number of pixels')

plt.show()
```

```
from skimage.exposure import match_histograms
from skimage import data
import matplotlib.pyplot as plt

# Load images
source = data.chelsea()
reference = data.rocket()

# Apply matching
matched = match_histograms(source, reference, channel_axis=-1)

# Show results
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.imshow(source)
plt.title('Source')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(reference)
plt.title('Reference')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(matched)
plt.title('Matched')
plt.axis('off')

plt.show()
```
